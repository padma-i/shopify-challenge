# Shopify Challenge - Question 1
This data explores the AOV question - finding issues with current approach, suggesting a new metric, and finding the new value.
There are comments to explain the process along the way, along with markdown textboxes.

```python
import pandas as pd
import numpy as np

#importing the data regarding the purchases and displaying it
data1 = pd.read_csv("2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv")
data1.head()
```

```python
#figuring out how the current AOV value of $3145.13 was found
sum(list(data1["order_amount"]))/len(list(data1["order_amount"]))
```

```python
# the AOV value of $3145.13 was calculated by dividing the sum of all the purchases by the number of orders
# thus, this data considered all purchases and orders including potential outliers.
# now, this code will graph the order_amounts as a boxplot with the outliers, giving insight into the spread of the data and outliers
import matplotlib.pyplot as plt
plt.boxplot(data1["order_amount"])
```

```python
#there were significant outliers that could have impacted the data. 
#now, this code also graphs the data without the outliers
plt.boxplot(data1["order_amount"], showfliers=False)
```

# Issues with Current Approach and New Suggested Metric

As can be seen in the boxplot above without outliers, the range for the 4th quartile is relatively larger than the range of the other quartiles. Looking at the two boxplots together, one can see that there are outliers and bigger purchases which lead to a inaccurate AOV value. Certain people may have bought shoes in bulk, impacting the mean and making it a poor metric to use.

Rather, using the metric of median will be more effective, since outliers will not impact it and thus it will be more representative of the data.

```python
#calculating the median
np.median(list(data1["order_amount"]))
```

# New Value
The new AOV value (using median as the metric) is $284.
